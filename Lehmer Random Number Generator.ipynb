How if random number generator that we always use is actually not random?

Well, the truth is bitter but it is actually not totally random. One of examples for random number generator is Lehmer algorithm. It is a systematical method to generate the "random" number. In this project we may see why and how to prove its unrandomness.

In this section, we will verify that a = 23 is indeed full-period, modulus compatible multipliers relative to m = 401.

```python
import math
import matplotlib.pyplot as plt
import numpy as np

m = 401
a = 23
multipliers=[]

x = [a]

for i in range (0, m-2):
    r = m % a
    q = m // a
    t = a * (x[i] % q) - r * (x[i] // q)
    if (t > 0):
        x.append(t)
    else:
        x.append(t + m)
y = np.roll(x, -1)
plt.figure()
plt.scatter(x, y, alpha=0.7)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("a = 23")
plt.show()
```

The plot has a pattern, this shows us that the "random" points are not random. 

Let us try another number for a!

### Section 2

Finally, if only 10 numbers generated by sequence are plotted, then things indeed look random. The illusion of randomness comes from generating few numbers (much less than m).

As we increasing the number of points that we plot, we will start seeing a lattice pattern (in this case) when we have around 100 numbers generated.

```python
m = 401
a = 66

# Define the numbers
test = [10, 20, 40, 100]

for j in test:
    multipliers=[]

    x = [a]

    for i in range (0,j):
        r = m % a
        q = m // a
        t = a * (x[i] % q) - r * (x[i] // q)
        if (t > 0):
            x.append(t)
        else:
            x.append(t + m)

    y = np.roll(x, -1)

    plt.figure()
    plt.scatter(x, y, alpha=0.85)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("a = 66 with total numbers= {0:2d} ".format(j))
    plt.show()
```
